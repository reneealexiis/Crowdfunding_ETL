```python
import pandas as pd
import json

df = pd.read_excel('Resources/contacts.xlsx')

df = df.iloc[3:]


all_values = []
for index, row in df.iterrows():
    row_dict = json.loads(row[0])
    values = [row_dict['contact_id'], row_dict['name'], row_dict['email']]
    all_values.append(values)

contact_info = pd.DataFrame(all_values, columns=['contact_id', 'name', 'email'])

contact_info
```

```python
contact_info.dtypes
```

```python
contact_info[['first_name', 'last_name']] = contact_info['name'].str.split(' ', n=1, expand=True)


contact_info.drop('name', axis=1, inplace=True)

contact_info = contact_info[['contact_id', 'first_name', 'last_name', 'email']]
contact_info 
```

```python

contact_info.dropna(inplace=True)

contact_info.drop_duplicates(inplace=True)

contact_info.to_csv('contacts.csv', index=False)


```
